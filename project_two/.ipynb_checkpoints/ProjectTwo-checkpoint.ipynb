## Question 3. Produce a bar plot of how many fossils exist for each subfamily. (10 pts). 

```python
fossils_counts = fossils.groupby(['subfamily'])['subfamily'].count()
fossils_counts
fossils_counts = fossils_counts.reset_index(name='counts')
fossils_counts
```

```python
fossils_counts
```

```python
(p9.ggplot(data=fossils_counts,
           mapping=p9.aes(x='subfamily',
                          y='counts',
                          color='subfamily'))
    + p9.geom_bar(stat='identity')
    + p9.theme_bw()
    + p9.theme(axis_text_x = p9.element_text(angle=90))
)
```
